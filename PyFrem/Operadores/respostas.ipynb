```python
# Faça um algoritmo que peça ao usuário para informar o preço de um produto e mostre seu novo preço, com 7% de desconto.

valor = float(input("Diga o valor do seu produto: "))

desconto = valor * 0.07
novoValor = valor - desconto

print(f"O valor do produto foi de: {valor} com um desconto de 7% vai para {novoValor}")
```

```python
# Faça um programa que calcule a área de um círculo a partir do raio informado pelo usuário. Imprima na tela o resultado.

raio = float(input("Informe o Raio do circulo: "))
pi = 3.14

area = pi * (raio ** 2)

print(f"""
O valor da área do círculo é de: {area}
""")
```

```python
# Faça um programa que calcule o valor total de uma compra. O usuário deve informar o preço de 3 produtos e a quantidade comprada de cada um. Imprima na tela o resultado.

valor1 = float(input("Preço do primeiro produto: "))
quant1 = int(input("Informe a quantidade do primeito produto: "))

valor2 = float(input("Preço do segundo produto: "))
quant2 = int(input("Informe a quantidade do Segundo: "))

valor3 = float(input("Preço do terceiro produto: "))
quant3 = int(input("A quantidade do terceiro produto: "))

preco1 = valor1 * quant1
preco2 = valor2 * quant2
preco3 = valor3 * quant3

print(f"""
O primeiro produto ficou com um valor de ({preco1})
O segundo produto ficou com ({preco2})
E o terceiro ({preco3})
""")
```

```python
# Escreva um programa para determinar se uma pessoa deve ou não pagar imposto. Considere que pagam imposto pessoa cujo salário é maior de R$ 1200,00. Peça para o usuário informar o valor do salário.

sala = float(input("Qual é seu salario: "))

if (sala > 1200.00):
  print(f"Com base no seu salario de ({sala}), sua pessoa é obrigada a pagar o imposto")
else:
  print(f"Você não precisa pagar o imposto")
```

```python
# Faça um programa que calcule o IMC(Índice de Massa Corporal) e mostre o resultado na tela. A fórmula é: peso/altura².

p = float(input("Informe seu peso atual: "))
a = float(input("Informe sua altura: "))

imc = p / (a ** 2)

print(f"Seu IMC é de: {imc}")
```
